```python
from crewai.knowledge.source.string_knowledge_source import StringKnowledgeSource

# Internal onboarding FAQ
faq_content = """
- You can access your email via portal.company.com using your employee credentials.
- The standard work hours are from 9am to 6pm, Monday to Friday.
- All reimbursement requests must be submitted by the 5th of the following month.
- For any IT-related issues, contact support@company.com.
"""

faq_knowledge_source = StringKnowledgeSource(
    content=faq_content,
    name="Internal Onboarding FAQ",
    description="A FAQ document for new employees covering common questions about company policies and procedures."
)

```

```python
from crewai import LLM, Agent, Crew, Task
llm = LLM(model="gpt-4o", temperature=0)
ollama_embedder = {
    "provider": "ollama",
     "config": {
        "model": "nomic-embed-text",
        "api_url": "http://localhost:11434"
    }
}

hr_faq_agent = Agent(
    name="HR FAQ Agent",
    role="You are an assistant that helps employees by answering questions about internal HR policies and procedures using the onboarding FAQ as your knowledge source.",
    goal ="Answer employee questions regarding HR policies and procedures based on the provided FAQ.",
    backstory="You are an internal HR assistant designed to help new employees navigate company policies and procedures effectively.",
    description="An agent that answers questions about internal HR policies and procedures using the onboarding FAQ.",
    llm=llm,
    knowledge_sources=[faq_knowledge_source],
    embedding_config=ollama_embedder
)

task = Task(
    name="Answer HR Policy Question",
    description="Answer the question: {question}",
    expected_output="The standard work hours are from 9am to 6pm, Monday to Friday.",
    agent=hr_faq_agent
)

crew = Crew(
    name="HR FAQ Crew",
    agents=[hr_faq_agent],
    tasks=[task],
    llm=llm,
    embedding_config=ollama_embedder,
    process="sequential",
    verbose=True
)

result = crew.kickoff(inputs={"question": "What are the standard work hours?"})
print(result)
```
